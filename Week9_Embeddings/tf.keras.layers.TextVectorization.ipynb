## tf.keras.layers.TextVectorization

### Explanation

`tf.keras.layers.TextVectorization` is a preprocessing layer used to convert raw text strings into a numerical format (integers or multi-hot encoding) that machine learning models can process. It performs four main steps:

1.  **Standardization**: Cleans the text (e.g., converting to lowercase, removing punctuation).
2.  **Tokenization**: Splits the strings into individual units (tokens), usually words.
3.  **Vocabulary Building**: Creates a mapping of unique tokens to integer indices.
4.  **Vectorization**: Converts each string into a sequence of integers (or a vector) based on the learned vocabulary.


### Code Example

This example demonstrates how to take raw sentences, build a vocabulary, and transform them into padded integer sequences.


```python
import tensorflow as tf

# 1. Sample Data
data = [
    "TensorFlow is amazing!",
    "Deep learning is a subset of machine learning.",
    "Text vectorization makes text numerical."
]

# 2. Initialize the layer
# standardize='lower_and_strip_punctuation' is the default
vectorize_layer = tf.keras.layers.TextVectorization(
    max_tokens=100,           # Maximum size of the vocabulary
    output_sequence_length=10 # Pad/truncate all sequences to length 10
)

# 3. "Adapt" the layer to the data
# This step scans the text to build the vocabulary (mapping words to IDs)
vectorize_layer.adapt(data)

# 4. Inspect the Vocabulary
vocab = vectorize_layer.get_vocabulary()
print("Vocabulary:", vocab)

# 5. Transform new text
test_samples = ["TensorFlow is great", "Machine learning text"]
vectorized_output = vectorize_layer(test_samples)

print("\nInput Samples:")
for i, sample in enumerate(test_samples):
    print(f"'{sample}'")

print("\nVectorized Output (Integer Sequences):")
print(vectorized_output.numpy())
```


### Key Parameters Breakdown

*   **`max_tokens`**: Limits the vocabulary size. Only the top $N$ most frequent words are kept; others are treated as `[UNK]` (unknown).
*   **`standardize`**: A function or string defining how to clean text. Default: `'lower_and_strip_punctuation'`.
*   **`split`**: How to break strings into tokens. Default: `'whitespace'`.
*   **`output_sequence_length`**: Ensures all outputs have the same shape. If a sentence is shorter, it is **padded** with zeros; if longer, it is **truncated**.
*   **`output_mode`**: 
    *   `'int'` (Default): Returns sequences of integers (ideal for RNNs/LSTMs).
    *   `'multi_hot'`: Returns a binary vector indicating presence/absence of words (ideal for simple Dense models).
    *   `'count'`: Returns the frequency of each word.

### Summary Table

| Feature | Input Type | Output Type | Use Case |
| :--- | :--- | :--- | :--- |
| **Integer Mode** | String | `[batch, seq_len]` (Integers) | RNNs, LSTMs, Transformers |
| **Multi-hot Mode** | String | `[batch, vocab_size]` (0 or 1) | Simple Sentiment Analysis |
| **Count Mode** | String | `[batch, vocab_size]` (Counts) | Bag-of-Words models |
