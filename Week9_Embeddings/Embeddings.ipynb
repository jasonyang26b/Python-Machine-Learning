### **Key words Definition:**
**One-hot encoding**
- A method of representing categrocial data as binary vectors, where a single index is set to 1 and all other indices are set to 0, creating a vector the size of the entire vocabular.

**Bag-of-words**
- A simplified text representation technique that counts the frequency of each word in a document, ignoring grammar, word order, and context.

**Tokenization:**
- The process of breaking down a stream of text into smaller, meaningful units called tokens, such as words, characters, or subwords.

**Word embeddings**
- Dense, low or high-dimensional vector representations of words where semantically similar words are positioned close to each other in a continuous vector space.

**Language Modeling**
- The task of predicting the probability of a sequence of words, typically by predicting the next word in a sentence given the preceding context.

**Sparse vs. Dense Vectors**
- Sparse vectors: Vectors containing mostly zeros, where most dimensions have no vlaue (common in one-hot encoding and bag-of-words)
- Dense vectors: Vectors where most dimensions contain non-zero values, representing compressed, feature rich information (common in word embeddings)

**Transfer learning**
- A machine learning technique where a model developed for one task is reused as the starting point for a model on a second, related task.

### **Differences between One-hot encoding, Bag-of-words, and Word Embeddings**
**`Input Dimensionality & Sparsity:`**
- **One-hot encoding & Bag-of-words (Sparse/high-dimensional)**: These methods create inputs where the vector size equals the vocabular size. In large datasets, the size can be hundreds of thousands. This results in sparse vectors (mostly zeros), which are computationally expensive to process and lead to the 'curse of dimensionality'.

- **Word Embeddings (Dense/low-Dimensional)**: These compress information into fixed, small dimensions. This reduces the memory footprint and allows the model to process much larger datasets more efficieently.

**`Semantic Learning:`**
- **One-hot encoding & Bag-of-words - lack of relationship**: In sparse representations, each word is equally distant from every other word.

- **Word Embeddings - keep semantic meaning:** Because embeddings map words to a continuous vector space based on context, the model starts with an inherent understanding of synonyms and analogies. Leaning is accelerated because the model can generalize: if it learns that 'excellent' is positive, it can mathematically apply that logic to 'good' because their vectors are close together.

### **Real Application of One-hot encoding, Bag-of-words, and Word Embeddings**
**For sentence: "I love coding."**

**Assume our total vocabular is: {"I","love", "coding","hate","bugs"}**

**`1) One-hot encoding:`**
- token: "I" -> [1, 0, 0, 0, 0]
- token: "love" -> [0, 1, 0, 0, 0]
- token: "coding" -> [0, 0, 1, 0, 0]

**Impact on Classification:** the model receives a sequence of vectors. It treats each word as an independent entity with no relationship to others.

**`2) Bag-of-word`**
- Sentence Vector:[1, 1, 1, 0, 0]

    - Meaning: (1 -> "I", 1 -> "love", 1 -> "coding", 0 -> "hate", 0 -> "bugs")

**Impact on Classification:** the model receives one fixed-length vector per sentence. This is highly efficient for simple task (like spam detection) where the presence of specifc words ('win', 'free', 'urgent') is more important than their order. However, it cannot distinguish between "I love coding" and "Coding love I".

**`3) Word Embeddings`**
- Instead of using fixed indices, each word is mapped to a dense vector of continuous numbers (below we use only 3-dimensions for simplicity). These values are learned during training.

- token: "I" -> [0.12, -0.59, 0.88]
- token: "love" -> [0.45, 0.22, -0.10]
- token: "coding" -> [0.78, -0.01, 0.34]

To represent the sentence for classification, we typically average these vectors to create a single "sentence embedding".

**Impact on Classification:** The model receives a dense, compressed representation. Because "love" and "like" would have similar vectors, the model can classify a sentence containing "I like coding" as "Positive" even if it has never seen the word "love" in a positive context before. This allows for much higher generalization.

### **Summary Evaluation Matrix**

| Feature | Sparse Encodings (One-hot Encoding & Bag-of-word) | Dense Encodings (Word-embedding) |
| :--- | :--- | :--- |
| **Best Use Case** | Simple keyword tasks (Spam, Topic ID) | Complex NLP (Translation, Summarization) |
| **Generalization** | Poor (Exact match only) | Excellent (Semantic similarity) |
| **Memory Usage** | High (due to large, empty vectors) | Low (compact, information-rich) |
| **Interpretability** | High (Traceable to specific words) | Low (Abstract mathematical features) |
| **Training Speed** | Fast (Simple counting/mapping) | Slow (Requires heavy computation) |
